```python
import face_recognition
import cv2
import matplotlib.pyplot as plt

# Load a sample picture and learn how to recognize it.
known_image = face_recognition.load_image_file(r"G:\My Drive\CS\Face_recognition (Matching)\Images\source_image.jpg") # Replace with your known image path
known_face_encoding = face_recognition.face_encodings(known_image)[0]

# Load an image to test
test_image = face_recognition.load_image_file(r"G:\My Drive\CS\Face_recognition (Matching)\Images\test-1.jpg") # Replace with your test image path
# Ensure the known image is in RGB format

# Find all the faces and face encodings in the test image
face_locations = face_recognition.face_locations(test_image)
face_encodings = face_recognition.face_encodings(test_image, face_locations)

# Convert the test image to BGR color (OpenCV uses BGR, not RGB)
test_image_bgr = cv2.cvtColor(test_image, cv2.COLOR_RGB2BGR)

for (top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
    # See if the face is a match for the known face(s)
    matches = face_recognition.compare_faces([known_face_encoding], face_encoding)

    if True in matches:
        # Draw a box around the face
        cv2.rectangle(test_image_bgr, (left, top), (right, bottom), (0, 255, 0), 2)
        cv2.putText(test_image_bgr, "Known Person", (left, top - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    else:
        cv2.rectangle(test_image_bgr, (left, top), (right, bottom), (0, 0, 255), 2)
        cv2.putText(test_image_bgr, "Unknown", (left, top - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)

# Convert the images to RGB format for displaying in matplotlib
known_image_rgb = cv2.cvtColor(known_image, cv2.COLOR_BGR2RGB)
test_image_rgb = cv2.cvtColor(test_image_bgr, cv2.COLOR_BGR2RGB)

# Create subplots to display both images side by side
plt.figure(figsize=(12, 6))

# Display the known image
plt.subplot(1, 2, 1)
plt.imshow(known_image_rgb)
plt.title("Known Image")
plt.axis('off')  # Hide axis

# Display the test image with face recognition results
plt.subplot(1, 2, 2)
plt.imshow(test_image_rgb)
plt.title("Test Image")
plt.axis('off')  # Hide axis

# Show the combined plot
plt.show()
```
